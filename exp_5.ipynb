```python
import socket

def resolve_domains(domains):
    """
    Takes a list of domain names and attempts to resolve them to IP addresses.
    Returns a dictionary mapping domains to their resolved IPs or an error message.
    """
    resolution_results = {}

    for domain in domains:
        # Strip potential URI schemes if accidentally included
        clean_domain = domain.replace("http://", "").replace("https://", "").split("/")[0]

        try:
            # gethostbyname_ex returns a tuple: (hostname, aliaslist, ipaddrlist)
            hostname, aliases, ip_addresses = socket.gethostbyname_ex(clean_domain)
            resolution_results[clean_domain] = ip_addresses
        except socket.gaierror:
            # Handles DNS resolution errors (e.g., domain doesn't exist)
            resolution_results[clean_domain] = ["Resolution Failed (NXDOMAIN or Network Error)"]

    return resolution_results

def main():
    # Target list for OSINT gathering
    target_domains = [
        "example.com",
        "google.com",
        "github.com",
        "invalid-domain-name-demo123.org"
    ]

    print("=== OSINT Infrastructure Mapping: DNS Resolution ===")
    print(f"Target Domains Loaded: {len(target_domains)}\n")

    # Perform resolution
    results = resolve_domains(target_domains)

    # Display the results
    for domain, ips in results.items():
        print(f"[>] Target: {domain}")
        for ip in ips:
            if "Failed" in ip:
                print(f"    [-] {ip}")
            else:
                print(f"    [+] Resolved IP: {ip}")
        print("-" * 40)

if __name__ == "__main__":
    main()
```
