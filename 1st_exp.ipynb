```python
import os
import time
from datetime import datetime

def print_metadata(filepath, state_name):
    stat = os.stat(filepath)

    print(f"\n[{state_name}]")
    print(f"File: {filepath}")
    print(f"Access Time (atime):       {datetime.fromtimestamp(stat.st_atime)}")
    print(f"Modification Time (mtime): {datetime.fromtimestamp(stat.st_mtime)}")
    print(f"Creation/Change (ctime):   {datetime.fromtimestamp(stat.st_ctime)}")
    print("-" * 50)

def main():
    filename = "forensic_evidence.txt"

    print("=== Locard's Exchange Principle: Digital Trace Demo ===")
    print("> Principle: 'Every contact leaves a trace.'")
    print("> We will create a file, wait, and then interact with it to observe the traces.\n")

    with open(filename, 'w') as f:
        f.write("Initial confidential data.\n")

    print_metadata(filename, "INITIAL STATE (Crime Scene Created)")

    print("Waiting for 2 seconds to simulate time passing...")
    time.sleep(2)

    print("\n>>> SUSPECT EVENT: Suspect finds the file and adds new data to it. <<<")
    with open(filename, 'a') as f:
        f.write("Suspect was here and modified the contents!\n")

    print_metadata(filename, "POST-EVENT STATE (Forensic Analysis)")

    print("=== Forensic Conclusion ===")
    print("Notice how the Modification Time (mtime) and Access Time (atime) have changed.")
    print("Even though the suspect only appended text, their interaction altered the file's metadata.")
    print("The contact left a digital trace, proving Locard's Exchange Principle!")

    if os.path.exists(filename):
        os.remove(filename)

if __name__ == "__main__":
    main()
```
