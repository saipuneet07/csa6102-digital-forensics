```python
import hashlib
import os

def calculate_sha256(filepath):
    """Reads a file in chunks and returns its SHA-256 hash."""
    sha256_hash = hashlib.sha256()

    # Read the file in chunks to handle large files efficiently
    with open(filepath, "rb") as f:
        for byte_block in iter(lambda: f.read(4096), b""):
            sha256_hash.update(byte_block)

    return sha256_hash.hexdigest()

def main():
    filename = "important_contract.txt"

    print("=== Data Integrity: Cryptographic Hash Demo ===")

    # Step 1: Create the original file
    with open(filename, "w") as f:
        f.write("Payment Authorized: $10,000 to Vendor A.")

    print(f"\n[1] Original file created: '{filename}'")

    # Step 2: Compute and store the initial hash (Baseline Integrity)
    original_hash = calculate_sha256(filename)
    print(f"    Original SHA-256 Hash: {original_hash}")

    # Step 3: Simulate file tampering (Malicious Actor modifies the file)
    print("\n[2] ALERT: Malicious actor is modifying the file...")
    with open(filename, "w") as f:
        f.write("Payment Authorized: $99,000 to Vendor A.")

    # Step 4: Compute the new hash after the file has been closed
    tampered_hash = calculate_sha256(filename)
    print(f"    New SHA-256 Hash:      {tampered_hash}")

    # Step 5: Compare the hashes to verify integrity
    print("\n=== Integrity Verification ===")
    if original_hash == tampered_hash:
        print("[SUCCESS] Hashes match. File integrity is intact.")
    else:
        print("[WARNING] Hashes DO NOT match! File has been tampered with.")

    # Cleanup
    if os.path.exists(filename):
        os.remove(filename)

if __name__ == "__main__":
    main()
```
