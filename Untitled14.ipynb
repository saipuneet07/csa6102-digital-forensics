```python
import re
from urllib.parse import urlparse

def check_phishing_url(url):
    score = 0
    reasons = []

    parsed_url = urlparse(url)
    domain = parsed_url.netloc

    if re.match(r"^\d{1,3}(\.\d{1,3}){3}$", domain.split(':')[0]):
        score += 3
        reasons.append("IP address used instead of domain name")

    if "@" in parsed_url.netloc or "@" in parsed_url.path:
        score += 3
        reasons.append("Contains '@' symbol")

    if len(url) > 75:
        score += 1
        reasons.append("URL length exceeds 75 characters")

    shorteners = ["bit.ly", "tinyurl.com", "goo.gl", "t.co", "is.gd", "ow.ly", "cutt.ly"]
    if any(shortener in domain for shortener in shorteners):
        score += 2
        reasons.append("Uses a known URL shortening service")

    if domain.count("-") > 1:
        score += 1
        reasons.append("Multiple hyphens detected in domain name")

    suspicious_keywords = ["login", "verify", "update", "secure", "account", "banking", "auth", "confirm", "wallet"]
    if any(keyword in url.lower() for keyword in suspicious_keywords):
        score += 2
        reasons.append("Contains suspicious keywords")

    if domain.count(".") > 3:
        score += 2
        reasons.append("Excessive subdomains detected")

    return score, reasons

def main():
    test_urls = [
        "https://www.example.com",
        "http://192.168.1.15/login.php",
        "https://bit.ly/3xY8Zq",
        "https://paypal-secure-update.com-login-verify.xyz/auth",
        "https://www.trustedbank.com@10.0.0.1/secure-update"
    ]

    for url in test_urls:
        print(f"\nAnalyzing URL: {url}")
        score, reasons = check_phishing_url(url)

        if score >= 3:
            print("Status: [!] HIGH RISK - Likely Phishing")
        elif score >= 1:
            print("Status: [-] MEDIUM RISK - Suspicious")
        else:
            print("Status: [+] LOW RISK")

        for reason in reasons:
            print(f" -> {reason}")

if __name__ == "__main__":
    main()
```
